$$
$$
\theta = arccos{( \frac{Z}{R} )}
$$
$$
\phi = arctan{( \frac{Y}{X} )}
$$
Normalized radius
$
\rho = {\frac{2R}{n} }$

$$
Y_{\ell}^{m}(\theta, \phi) = \sqrt{\frac{2n+1}{4\pi}\frac{(n-m) !}{(n+m) !}}e^{im\theta}P_{\ell}^{m}cos(\phi)
$$
$$
\Psi_{nlm}(X, Y, Z) = \sqrt{ \frac{8}{n^3} \frac{(n - l - 1)!}{2n (n + l)!}}e^{-\rho/2}{L_{n - l - 1}^{2l + 1}(\rho)}\rho^{l}Y_{\ell}^{m}(\theta, \phi)
$$



```python
import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider, Button, RadioButtons
import scipy.special
from scipy.special import sph_harm
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.colors import ListedColormap


def hydrogen_wf(n,l,m,X,Y,Z):
    R = np.sqrt(X**2+Y**2+Z**2)
    Theta = np.arccos(Z/R)
    Phi = np.arctan2(Y,X)
    
    rho = 2.*R/n
    s_harm=sph_harm(m, l, Phi, Theta)
    l_poly = scipy.special.genlaguerre(n-l-1,2*l+1)(rho)
    
    prefactor = numpy.sqrt((2./n)**3*math.factorial(n-l-1)/(2.*n*math.factorial(n+l)))
    wf = prefactor*numpy.exp(-rho/2.)*rho**l*s_harm*l_poly
    wf = np.nan_to_num(wf)
    return wf
  
dz=0.15
zmin=-10
zmax=10
x = np.arange(zmin,zmax,dz)
y = np.arange(zmin,zmax,dz)
z = np.arange(zmin,zmax,dz)
X,Y,Z = np.meshgrid(x,y,z) #X, Y, Z are 3d arrays that tell us the values of x, y, and z at every point in space
```

For the radial probability distribution we need to integrate the probability density over the angular components. The radial probability distribution $P(r)$ for a hydrogen atom is given by:

$$P(r)=r^2∣R_{nl}(r)∣^2$$

where $R_{nl}(r)$ is the radial part of the wavefunction.

```python

def hydrogen_radial_wf(n, l, r):
    rho = 2 * r / n
    l_poly = scipy.special.genlaguerre(n - l - 1, 2 * l + 1)(rho)
    prefactor = np.sqrt((2. / n)**3 * math.factorial(n - l - 1) / (2. * n * math.factorial(n + l)))
    radial_wf = prefactor * np.exp(-rho / 2.) * rho**l * l_poly
    return radial_wf

# Define the quantum numbers for the orbitals
orbitals = [
    (1, 0, 0),  # 1s
    (2, 0, 0),  # 2s
    (2, 1, 1),  # 2px
    (2, 1, -1), # 2py
    (2, 1, 0)   # 2pz
]

r = np.linspace(0, 20, 1000)

fig, ax = plt.subplots()

for n, l, m in orbitals:
    radial_wf = hydrogen_radial_wf(n, l, r)
    radial_prob = r**2 * np.abs(radial_wf)**2
    ax.plot(r, radial_prob, label=f'n={n}, l={l}, m={m}')

ax.set_xlabel('r')
ax.set_ylabel('Radial Probability Density')
ax.set_title('Radial Probability Distribution of Hydrogen Atom Orbitals')
ax.legend()
plt.show()
```

* 1s Orbital (n=1, l=0, m=0)

    Radial Probability Distribution: In accordo con la teoria di Bohr, la quale afferma che è più probabile trovare gli elettroni a una specifica distanza dal nucleo. Non ci sono nodi (punti dove la densità di probabilità è nulla)

* 2s Orbital (n=2, l=0, m=0)

    Radial Probability Distribution: C'è un nodo: in una certa distanza (non lontana dal nucleo) la probabilità di trovare un elettrone è nulla
  
* 2p Orbitals (n=2, l=1, m=-1, 0, 1)

    Radial Probability Distribution: tutti gli orbitali 2p (2px, 2py, 2pz) hanno distribuzione radiale con massimo a una distanza più lontana dei 2s e senza nodi.

```python
a0_real = 1  # Bohr radius in meters

def R_2s_real(r):
    return (1/(2*np.sqrt(2))) * (1/a0_real)**(3/2) * (2 - r / a0_real) * np.exp(-r / (2 * a0_real))

def Y_2s(theta, phi):
    return np.sqrt(1 / (4 * np.pi))

# Setting up the 3D grid
x_real = np.linspace(-5 * a0_real, 5 * a0_real, 500)
y_real = np.linspace(-5 * a0_real, 5 * a0_real, 500)
X_real, Y_real = np.meshgrid(x_real, y_real)
Z_real = np.zeros_like(X_real)

r_real = np.sqrt(X_real**2 + Y_real**2)
theta_real = np.arctan2(np.sqrt(X_real**2 + Y_real**2), Z_real)
phi_real   = np.arctan2(Y_real, X_real)

fig_real = plt.figure(figsize=(9, 6))
ax_real  = fig_real.add_subplot(111, projection='3d')

# 2s Orbital
psi_val_real = np.abs(R_2s_real(r_real) * Y_2s(theta_real, phi_real))**2
ax_real.plot_surface(X_real, Y_real, psi_val_real, cmap='cividis')
ax_real.set_title('2s Orbital')

plt.show()
```

### Average distance of an electron in a 1s, 2s, 2px, 2py, and 2pz orbital from the nuclei.


## 1s
$$<r> = \int_0^\infty r \frac{\partial p}{\partial r} dr = \frac{4}{a^3_0} \int_0^\infty r^3 e^{-2r/a_0} dr  $$

Integrazione per parti, dato che s ha simmetria sferica, la probabilità dipende solo da r:
$$\frac{4}{a^3_0} [ e^{-2r/a_0}(\frac{-a_0r^3}{2} - \frac{3a^2_0r^2}{4} - \frac{3a^3_0r}{4} - \frac{3a^4_0}{8})]^\infty_0$$
$$=\frac{3}{2}a_0$$


## 2s
$$<r> = \frac{a_0}{2}(3n^2 - l(l+1)) = 6a_0$$

Più mi allontano dal nucleo, più aumenta l'energia, più aumenta il valor medio della distanza: siccome la forza attrattiva tra elettroni e nucleo (a carica positiva) aumenta col numero atomico $\, \rightarrow \,$ elettrone attratto dal nucleo. 

## 2p
Per gli orbitali p, è diverso andare in una direzione invece che un'altra: non sono simmetrici.
Dobbiamo per forza fare gli integrali tripli.
Il valor medio della distanza per i 2p è uguale (2px=2py=2pz).
$$\psi = \frac{1}{4\sqrt{2\pi}}\frac{1}{a^{3/2}_0}\frac{r}{a_0}e^{\frac{r}{2a_0}}\cos \theta $$
$$<r> = <\psi|r|\psi> = \int_0^\infty dr \int_0^\pi d \theta \int_0^{2\pi} d \phi r \psi^2 r^2 sin \theta$$
$$\int_0^\infty \int_0^\pi \int_0^{2\pi} \frac{1}{32 \pi a^5_0} r^2 e^{r^2/a_0} cos^2 \theta r^3 sin \theta \, dr \,d \theta \,d \phi $$
$$\int_0^\infty \frac{1}{24  a^5_0} r^5 e^{-r}{a_0} \,dr \quad  = \quad \frac{120a^6_0}{24a^5_0} \,=\, 5a_0$$
