Il MEP (Minimum Energy Pathway) è il percorso lungo il quale si minimizza l'energia durante la trasformazione da metano a gruppo metile e idrogeno.


Decresce per la formazione del transition state e poi aumenta per lo stabilizzarsi dei prodotti. Sorpassa la barriera energetica.

A distanza di circa 1.1 A.U. si ha il minimo (transition state).

$d_{C-H} \rightarrow \infty$, $E \rightarrow E_d $.

HF sovrastima l'energia di dissociazione, perché non considerando la correlazione elettronica il sistema sembri meno stabile di quanto non sia realmente.
CASPT2 inizia con una base di Complete Active Space (CAS) che permette una descrizione accurata degli stati elettronici più importanti, e poi utilizza la teoria delle perturbazioni di secondo ordine per includere gli effetti della correlazione elettronica dinamica. Questo correggere le sovrastime energetiche fatte da HF.

Il metodo CASPT2 si basa su un approccio a due stadi:
* Spazio Attivo Completo (CAS): trova gli orbitali molecolari che sono cruciali per la descrizione dello stato elettronico. Questo sottospazio, detto spazio attivo, include di solito i livelli energetici importanti per la reattività chimica, come gli orbitali di anti-legame coinvolti nella rottura dei legami. Calcola quindi le funzioni d'onda considerando le configurazioni dentro gli orbitali (CAS-SCF: sceglie tutti gli orbitali eccitabili, cioè sotto gli HOMO e portarli in tutti quelli sopra i LUMO) (un Full CI in uno spazio ridotto però). 
* Teoria delle Perturbazioni di Secondo Ordine (PT2): Una volta trovato il CAS, con questa teoria si includono gli effetti della correlazione elettronica. Considera le eccitazioni al di fuori dello spazio attivo, correggendo le energie calcolate nel primo passo.


Quando si rompe il legame C-H gli elettroni del legame e quelli circostanti interagiscono. HF non può catturare queste interazioni dinamiche in modo accurato, perché considera solo un campo medio statico. 
Col CASPT2 invece, si considerano sia le interazioni statiche all'interno del CAS, sia le interazioni dinamiche con gli stati elettronici esterni.


```python
from scipy.optimize import curve_fit

def morse_potential(r, D_e, a, r_e):
    return D_e * (np.exp(-2 * a * (r - r_e)) - 2 * np.exp(-a * (r - r_e)))
```

```python
distances = np.array(ch4_scan["DISTS"])
energies  = np.array(ch4_scan["ERS2"])
param_guess = [1.0, 1.0, 2.0]
energies = energies - energies[-1]
# Fit Morse potential to the shifted data
potential, _ = curve_fit(morse_potential, distances, energies, p0=param_guess)
```

```python
# Generate new distance values for interpolation
new_distances = np.linspace(min(distances), max(distances), 50)

fitted_energy = morse_potential(new_distances, *potential)


plt.plot(distances, energies, 'o', label='CASPT2 energy')
plt.plot(new_distances, fitted_energy, '-', label='Fitted Morse Potential')
plt.xlabel('Distance (r)')
plt.ylabel('Energy')
plt.legend()
plt.title('Morse Potential Fit')

plt.tight_layout()
plt.show()

# Print the fitted parameters
print("More parameters: (D_e, a, r_e):", potential)
```
